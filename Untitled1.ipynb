```python
import folium

# ตำแหน่งที่ต้องการแสดงบนแผนที่ (ละติจูดและลองจิจูดของแต่ละตำแหน่ง)
locations = [(9.153269586, 98.79568043), (6.42214196, 101.2731991), (7.657735823, 99.85935491)]

# สร้างแผนที่
m = folium.Map(location=[13.7563, 100.5018], zoom_start=10)

# เพิ่มหมุดบนแผนที่สำหรับแต่ละตำแหน่ง
for location in locations:
    folium.Marker(location, popup='ตำแหน่งCY').add_to(m)

# แสดงแผนที่
m
```
